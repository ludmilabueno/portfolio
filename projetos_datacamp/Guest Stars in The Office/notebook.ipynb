```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

## Importing dataset

```python
data = pd.read_csv('datasets/office_episodes.csv')
```

## Visualizing dataset

```python
data.head()
```

## Visualization of the episode number versus viewership (in million)

```python
# Defining color scheme
colors = []

for lab, row in data.iterrows():
    #print(lab, row)
    if row["scaled_ratings"] < 0.25:
        colors.append("red")
    elif row["scaled_ratings"] >= 0.25 and row["scaled_ratings"] < 0.5:
        colors.append("orange")
    elif row["scaled_ratings"] >= 0.5 and row["scaled_ratings"] < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
```

```python
# Defining sizing system
size = []

for lab, row in data.iterrows():
    if row["has_guests"] == True:
        size.append(250)
    else:
        size.append(25)
```

```python
# Adding columns in the DataFrame with the colors and size
data["colors"] = colors
data["size"] = size
```

```python
# Spliting the DataFrame into one with episodes with guest stars and other without guest stars
ep_with_guests = data[data["has_guests"] == True]
ep_with_no_guests = data[data["has_guests"] == False]
```

```python
# Parameter of figure size and plot style
plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')
```

```python
fig = plt.figure()
plt.scatter(x=ep_with_guests["episode_number"], y=ep_with_guests["viewership_mil"], marker='*', s=250, c=ep_with_guests["colors"])
plt.scatter(x=ep_with_no_guests["episode_number"], y=ep_with_no_guests["viewership_mil"], s=25, c=ep_with_no_guests["colors"])
plt.title("Popularity, Quality, and Guest Appearances on the Office", fontsize=20)
plt.xlabel("Episode Number", fontsize=16)
plt.ylabel("Viership (Millions)", fontsize=16)
plt.show()
```
